```python
for txt in academic_texts:
  training_data.append((txt, {'cats':{'REFERENCE':0.0, 'NORMAL': 1.0}}))
```

```python
from sklearn.model_selection import train_test_split

train_data, valid_data = train_test_split(training_data, test_size=0.2, random_state=42)
```

```python
import spacy
from spacy.training.example import Example

nlp = spacy.blank("en")
textcat = nlp.add_pipe("textcat")
textcat.add_label("REFERENCE")
textcat.add_label("NORMAL")
loss_list = []
optimizer = nlp.begin_training()
for i in range(20):
    losses = {}
    for text, annotations in train_data:
        example = Example.from_dict(nlp.make_doc(text), annotations)
        nlp.update([example], sgd=optimizer, losses=losses)
    print(losses)
    loss_list.append(losses)
```
